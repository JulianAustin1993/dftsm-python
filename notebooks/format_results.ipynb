# Format Results

In this notebook we format our result obtained in both our simulated and real world experiment. We display these results in pandas data frame. 

## Simulated Results
In the following code block we load our result files for the simulated experiments and summarise them for each of the noise types and forecasting steps. 

```python
PATH = '../results/simulated'
model =['fpca','mafr','reg-fpca', 'reg-mafr']
noise =['ln', 'hn','sn']
res_dict = {'_forecast':[1,3,10,15,20,25]}
for m in model:
    for n in noise:
        data = np.load(PATH+'_'+m+'_'+n+'.npz')
        mean =np.mean(data['results'], axis=0).round(4)
        sd = np.std(data['results'], axis=0).round(3)
        res_dict[m+'_'+n]=list(zip(mean,sd))
df = pd.DataFrame(res_dict)
df.set_index('_forecast')
df_cols = [c.split('_') for c in df.columns]
df.columns = pd.MultiIndex.from_tuples([(c[0], c[1]) for c in df_cols])
df = df.T
df
```
